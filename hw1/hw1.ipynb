```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

n = 1000 
p = 0.7  # Probability of Xt = 1
q = 0.3  # Probability of Xt = -1

Xt = np.random.choice([1, -1], size=n, p=[p, q])
St = np.cumsum(Xt)
df = pd.DataFrame({'t': range(n), 'Xt': Xt, 'St': St})
```

```python
plt.plot(range(n), St)
plt.xlabel('Time (t)')
plt.ylabel('St')
plt.title('Time Series of St')
plt.show()
```

```python
k1 = 30 
k2 = 365


df['ma_short'] = df['St'].rolling(window=k1).mean()
df['ma_long'] = df['St'].rolling(window=k2).mean()

# Create plots of the smoothed trends
plt.figure(figsize=(7, 5))
plt.plot(df['t'], df['ma_short'], label=f'Short-Term (k={k1})')

plt.xlabel('Time (t)')
plt.ylabel('Smoothed St')
plt.title('Smoothed Trends with MA Filters')
plt.legend()
```

```python
plt.figure(figsize=(7, 5))
plt.plot(df['t'], df['ma_long'], label=f'Long-Term (k={k2})')

plt.xlabel('Time (t)')
plt.ylabel('Smoothed St')
plt.title('Smoothed Trends with MA Filters')
plt.legend()
```

```python
df = pd.read_csv('./data/airline-passengers.csv')
df
```

```python
df['Month'] = pd.to_datetime(df['Month'])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Passengers'], label='Passengers')
plt.xlabel('Month')
plt.ylabel('Number of Passengers (in thousands)')
plt.title('Airline Passengers Time Series')
plt.grid(True)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plt.figure(figsize=(10, 5))
plot_acf(df['Passengers'], lags=12)
plt.title('Autocorrelation Function (ACF) Plot')
plt.grid(True)
plt.show()
```
